Using the requests and beautifullSoup libraries to scrape the names, ranks, country, and scores from Times Higher Education University ranking website

```python
# acquisition.ipynb
# This notebook acquires the 2025 CWUR university rankings and saves it as a CSV.

# Import necessary libraries
import requests
import pandas as pd

# Define the URL for the CWUR 2025 rankings
url = "https://cwur.org/2025.php"

# Set headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# Make the HTTP request
response = requests.get(url, headers=headers)
response.raise_for_status()  # Raise an error if the request failed

# Read all HTML tables from the page
tables = pd.read_html(response.text)

# Select the first table (main rankings table)
df = tables[0]

# Save the table to a CSV file
df.to_csv("cwur_rankings_2025.csv", index=False)
print("Data saved successfully as 'cwur_rankings_2025.csv'")
```
